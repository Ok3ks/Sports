## UTILS 

```python
import json

def to_json(obj,fp):
    """obj is the object, fp is the filepath"""
    with open(fp, 'w') as out:
        json.dump(obj, out)

def from_json(fp):
    with open(fp, 'w') as ins:
        obj = json.load(ins)
    
    return obj
```
